```python
import altair as alt
from altair import datum
import ipywidgets as widgets
from ipywidgets import interact, interact_manual
from IPython.display import HTML, Markdown as md
from tabulate import tabulate

from src.data.read_dataset import get_processed_dataset
from src.visualization import visualize as viz

ecdc = get_processed_dataset('ECDC')
```

```python
country = 'Austria'
```

```python
country_df = ecdc[ecdc.Country == country]
today = country_df.iloc[0]
yesterday = country_df.iloc[1]
```

```python
rows = [
        ('Total cases', today.TotalCases, today.TotalCases - yesterday.TotalCases),
        ('Total deaths', today.TotalDeaths, today.TotalDeaths - yesterday.TotalDeaths),
        ('Lethality', today.Lethality, ),
    ]

html = '<p>Last update: ' + str(today.dateRep) + '</p>'
html += tabulate(rows, headers=["", "Today", "Variation"], tablefmt="html")
HTML(html)
```

```python
from altair import datum

base = alt.Chart(country_df).mark_line().encode(
    x='Date:T',
    tooltip = ['Date', 'TotalCases', 'TotalDeaths', 'cases', 'deaths']
).properties(
    height=600,
    width=800
).transform_filter(
    (datum.TotalCases > 1000)
)
```

```python
c1 = base.encode(
    y=alt.Y('TotalCases', scale=alt.Scale(type='log'))
)

c2 = base.encode(
    y=alt.Y('TotalDeaths', scale=alt.Scale(type='log')),
    color=alt.value("#FF2200")
)

alt.layer(c1, c2).interactive()
```
