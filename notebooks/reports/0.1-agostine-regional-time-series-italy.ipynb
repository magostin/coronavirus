```python
import altair as alt

from src.data.read_dataset import get_processed_dataset
from src.visualization import visualize
```

```python
regions_df = get_processed_dataset('DPC')
```

## Number of confirmed cases by region

```python
alt.themes.enable('theme_1')
from altair import datum

alt.Chart(regions_df).mark_line().encode(
    alt.Y('totali', scale=alt.Scale(type='log')),
    x='date',
    color='regione',
    tooltip=['regione', 'date', 'totali']
).properties(
    height=600,
    width=600
).transform_filter(
    (datum.totali > 20)
).interactive()
```

## Number of confirmed cases by region, stacked

```python
alt.Chart(regions_df).mark_area().encode(
    alt.Y('totali'),#, scale=alt.Scale(type='log')),
    x='date',
    color='regione',
    tooltip=['regione', 'date', 'totali']
).properties(
    height=600,
    width=600
).interactive()
```

## Number of deaths by region

```python
alt.Chart(regions_df).mark_area().encode(
    alt.Y('deceduti'),
    x='date',
    color='regione',
    tooltip=['regione', 'date', 'deceduti']
).properties(
    height=600,
    width=600
).transform_filter(
    (datum.deceduti > 0)
).interactive()
```
